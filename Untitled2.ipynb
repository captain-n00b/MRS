```
#'Cough','Chest pain','Loose motion','pain at heart','Nausea  Vomiting','Difficulty Breathing','Headache','Stomachache','Amenia','Bleeding','Bloating','Constipation','Numbness','Fatigue','Fever'
from sklearn.preprocessing import LabelEncoder
le_Diseases=LabelEncoder()
le_Cough=LabelEncoder()
le_Chestpain=LabelEncoder()
le_Loosemotion=LabelEncoder()
le_painatheart=LabelEncoder()
le_Nausea=LabelEncoder()
le_Vomiting=LabelEncoder()
le_DifficultyBreathing=LabelEncoder()
le_Headache=LabelEncoder()
le_Stomachache=LabelEncoder()
le_Amenia=LabelEncoder()
le_Bleeding=LabelEncoder()
le_Bloating=LabelEncoder()
le_Constipation=LabelEncoder()
le_Numbness=LabelEncoder()
le_Fatigue=LabelEncoder()
le_Fever=LabelEncoder()
df['diseases']=le_Diseases.fit_transform(df['Diseases'])

df['cough']=le_Cough.fit_transform(df['Cough'])
df['chest pain']=le_Chestpain.fit_transform(df['Chest pain'])
df['loose motion']=le_Loosemotion.fit_transform(df['Loose motion'])
df['pain at heart']=le_Loosemotion.fit_transform(df['pain at heart'])
df['nausea']=le_Loosemotion.fit_transform(df['Nausea'])
df['vomiting']=le_Loosemotion.fit_transform(df['Vomiting'])
df['difficulty breathing']=le_Loosemotion.fit_transform(df['Difficulty Breathing'])
df['headache']=le_Loosemotion.fit_transform(df['Headache'])
df['stomachache']=le_Loosemotion.fit_transform(df['Stomachache'])
df['amenia']=le_Loosemotion.fit_transform(df['Amenia'])
df['bleeding']=le_Loosemotion.fit_transform(df['Bleeding'])
df['bloating']=le_Loosemotion.fit_transform(df['Bloating'])
df['constipation']=le_Loosemotion.fit_transform(df['Constipation'])
df['numbness']=le_Loosemotion.fit_transform(df['Numbness'])
df['fatigue']=le_Loosemotion.fit_transform(df['Fatigue'])
df['fever']=le_Loosemotion.fit_transform(df['Fever'])
inputs=df.drop(['cough','chest pain','loose motion','pain at heart','nausea','vomiting','difficulty breathing','headache','stomachache','amenia','bleeding','bloating','constipation','numbness','fatigue','fever'],axis='columns')
inputs.head()
target=inputs['diseases']
print(target)
```
